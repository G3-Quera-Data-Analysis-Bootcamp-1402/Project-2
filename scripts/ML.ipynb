```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pyp
```

# Q1

```python
fact_df = pd.read_csv('../data/fact.csv', delimiter='\t')
fact_df.head(5)
```

```python
product_df = pd.read_csv("../data/cleanized/product_dim.csv")
product_df.head(5)
```

```python
order_df = pd.read_csv("../data/cleanized/order_dim.csv")
order_df.head(5)
```

```python
geo_df = pd.read_csv("../data/cleanized/geo_dim.csv")
geo_df.head(5)
```

```python
df = fact_df.set_index("Product ID", drop= False).join(product_df.set_index("Product ID", drop= False), rsuffix= "_product")\
              .set_index("Order ID", drop= False).join(order_df.set_index("Order ID", drop= False), rsuffix= "_order")\
              .set_index("Geo ID", drop= False).join(geo_df.set_index("Geo ID", drop= False), rsuffix = "_geo")
df.head(5)
```

```python
X = df.drop(['Profit', 'Row ID', 'Order ID', 'Product ID', 'Customer ID', 'Geo ID', 'Date ID',\
    'Product ID_product', 'Product Name', 'Order ID_order', 'Geo ID_geo', 'City', 'State', 'Region'], axis=1)
X
```

```python
y = np.array(df["Profit"])
y
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from flaml import AutoML
from sklearn.decomposition import PCA
```

```python
# Standardization and Normalization
min_max_scaler = MinMaxScaler()

min_max_scaler.fit(X)
min_max_scaled = min_max_scaler.transform(X)
min_max_scaled_df = pd.DataFrame(min_max_scaled, columns=X.columns)

std_scaler = StandardScaler()

std_scaler.fit(min_max_scaled_df)
scaled = std_scaler.transform(min_max_scaled_df)
scaled_df = pd.DataFrame(scaled, columns= min_max_scaled_df.columns)
```

```python
# PCA transformation
pca_transformer = PCA(n_components=0.95, svd_solver = 'auto')
pca_transformer.fit(scaled_df)
reduced_X = pca_transformer.transform(scaled_df)
reduced_X
```

```python
reduced_X.shape
```

```python
# Without PCA
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=313)
```

```python
lrmodel = LinearRegression()
lrmodel.fit(x_train, y_train)
y_pred = lrmodel.predict(x_test)
r2_score(y_test, y_pred)
```

```python
amlmodel = AutoML(task= "regression", time_budget= 60, verbose=0)
amlmodel.fit(x_train, y_train)
y_pred = amlmodel.predict(x_test)
r2_score(y_test, y_pred)
```

```python
# With PCA
x_train, x_test, y_train, y_test = train_test_split(reduced_X,y, test_size=.2, random_state=313)
```

```python
lrmodel = LinearRegression()
lrmodel.fit(x_train, y_train)
y_pred = lrmodel.predict(x_test)
r2_score(y_test, y_pred)
```

```python
amlmodel = AutoML(task= "regression", time_budget= 60, verbose=0)
amlmodel.fit(x_train, y_train)
y_pred = amlmodel.predict(x_test)
r2_score(y_test, y_pred)
```
