# Importing Liberaries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pyp
```

# Data

```python
fact_df = pd.read_csv('../data/fact.csv', delimiter='\t')
fact_df.head(5)
```

```python
product_df = pd.read_csv("../data/cleanized/product_dim.csv")
product_df.head(5)
```

```python
order_df = pd.read_csv("../data/cleanized/order_dim.csv")
order_df.head(5)
```

```python
geo_df = pd.read_csv("../data/cleanized/geo_dim.csv")
geo_df.head(5)
```

```python
customer_df = pd.read_csv("../data/cleanized/customer_dim.csv")
customer_df.head(5)
```

# Preprocessing Finlization

## DF Works

```python
df = fact_df.set_index("Product ID", drop= False).join(product_df.set_index("Product ID", drop= False), rsuffix= "_product")\
              .set_index("Order ID", drop= False).join(order_df.set_index("Order ID", drop= False), rsuffix= "_order")\
              .set_index("Geo ID", drop= False).join(geo_df.set_index("Geo ID", drop= False), rsuffix = "_geo")\
              .set_index("Customer ID", drop= False).join(customer_df.set_index("Customer ID", drop= False), rsuffix= "_customer")
df.head(5)
```

```python
df.columns.tolist()
```

```python
X = df.drop(['Profit', 'Row ID', 'Order ID', 'Product ID', 'Customer ID', 'Geo ID', 'Date ID',\
    'Product ID_product', 'Product Name', 'Order ID_order', 'Geo ID_geo', 'City', 'State',\
    'Customer ID_customer', 'Customer Name',], axis=1)
X
```

```python
X.info()
```

```python
y = np.array(df["Profit"])
y
```

```python
# Importing some other Moduls
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from flaml import AutoML
from sklearn.decomposition import PCA
```

## Scaling and Dimention Reduction by PCA

```python
# Standardization and Normalization
min_max_scaler = MinMaxScaler()

min_max_scaler.fit(X)
min_max_scaled = min_max_scaler.transform(X)
min_max_scaled_df = pd.DataFrame(min_max_scaled, columns=X.columns)

std_scaler = StandardScaler()

std_scaler.fit(min_max_scaled_df)
scaled = std_scaler.transform(min_max_scaled_df)
scaled_df = pd.DataFrame(scaled, columns= min_max_scaled_df.columns)
```

```python
# PCA transformation
pca_transformer = PCA(n_components=0.99, svd_solver = 'auto')
pca_transformer.fit(scaled_df)
reduced_X = pca_transformer.transform(scaled_df)
reduced_X
```

```python
reduced_X.shape
```

# Modelling And Prediction

## Without PCA

```python
# Without PCA
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=128)
```

```python
lrmodel = LinearRegression()
lrmodel.fit(x_train, y_train)
y_pred = lrmodel.predict(x_test)
y_pred_lrmodel = y_pred
r2_score(y_test, y_pred)
```

```python
amlmodel = AutoML(task= "regression", time_budget= 300, verbose=0)
amlmodel.fit(x_train, y_train)
y_pred = amlmodel.predict(x_test)
r2_score(y_test, y_pred)
```

```python
d = {'y_test': y_test, 'y_pred_automl': y_pred, 'y_pred_lrmodel': y_pred_lrmodel}
df_test_pred = pd.DataFrame(d).reset_index(drop=True)
df_test_pred.to_csv('../data/df_ML1.csv')
```

## Modelling using MLP

```python
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.Dense(units=125))#units=5, input_shape=(x_train, x_train.shape[2])))

model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.Dense(units=1))

model.compile(loss='mae', optimizer='adam')
```

```python
x_train.head(5)
```

```python
x_train = x_train.astype({'Sales': 'float32', 'Quantity': 'float32', 'Discount': 'float32', 'Shipping Cost': 'float32'})
```

```python
x_train
```

```python
history = model.fit(x_train.iloc[:, 2:3], y_train,epochs=35, batch_size=32,shuffle=True)
```

```python
model.summary()
```

```python
y_pred = model.predict(x_test.iloc[:, 2:3])

r2_score(y_test, y_pred)
```
