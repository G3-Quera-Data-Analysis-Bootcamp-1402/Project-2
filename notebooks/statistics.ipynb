# Discount effects on sales, quantity, profit of products

```python
import os
import pathlib

from superstore.statistics.discount import divide_data, plot_data, ttest

data_dir = pathlib.Path("../data/").resolve()
```

## Split data into 2 groups: Discounted, Undiscounted

```python
divide_data(data_dir)
```

## Perfom T-test to compare 3-metrics (Sales, Quantity, Profit) in groups (Discounted, Undiscounted)

```python
keys = ["Sales", "Quantity", "Profit"]
t_statistics, p_values = ttest(data_dir)
for t_s, p_v, key in zip(t_statistics, p_values, keys):
    print(f"{key} - t-statistics: {t_s}, p-value: {p_v}")
```

## Visualize the differences

```python
plot_data(data_dir)
```
