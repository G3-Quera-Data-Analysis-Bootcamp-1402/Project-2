# Discount effects on sales, quantity, profit of products

```python
import pathlib

from superstore.statistics.discount import divide_data, plot_data, ttest_data, store_data

data_dir = pathlib.Path("../data/").resolve()
```

## Split data into 2 groups: Discounted, Undiscounted

```python
d_df, und_df = divide_data(data_dir)
print(d_df, und_df)
```

## Perfom T-test to compare 3-metrics (Sales, Quantity, Profit) in groups (Discounted, Undiscounted)

```python
ttest_df = ttest_data(d_df, und_df)
print(ttest_df)
```

## Visualize the differences

```python
plot_data(d_df, und_df)
```

## Store data in CSV files

```python
store_data([d_df, und_df, ttest_df], ["discounted", "undiscounted", "ttest"], [False, False, True], data_dir)
```
